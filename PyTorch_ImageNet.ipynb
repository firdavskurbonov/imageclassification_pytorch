```python

import torch
import torchvision.models as models
import torchvision.transforms as transforms
import torch.nn.functional as F
from PIL import Image

# Load the ResNet model with pre-trained weights
resnet = models.wide_resnet101_2(pretrained=True)

# Set the model to evaluation mode
resnet.eval()

```

```python
from torchvision.datasets import ImageFolder

# Define the data directory
data_dir = 'data'
```

```python
# Preprocess the image
preprocess = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406],
                         std=[0.229, 0.224, 0.225])
])
```

```python
# Load the dataset using ImageFolder
dataset = ImageFolder(root=data_dir, transform=preprocess)
```

```python
from collections import defaultdict
output = defaultdict(set)
```

```python
# Iterate through the dataset
for inputs, labels in dataset:
    # Add a batch dimension to the inputs tensor
    inputs = inputs.unsqueeze(0)

    # Pass the inputs through the model
    outputs = resnet(inputs)

    # Convert the outputs to probabilities using softmax function
    probs = F.softmax(outputs, dim=1)

    # Get the predicted class index (the class with highest probability)
    pred_index = probs.argmax(dim=1)

    # Get the predicted class label using the ImageNet labels file
    with open('imagenet_classes.txt') as f:
        labels_list = f.readlines()
    pred_label = labels_list[pred_index.item()]

    # Print the predicted class label and actual class label
    output[dataset.classes[labels]].add(pred_label)
    print('Predicted: {}'.format(pred_label))
    print('Actual: {}'.format(dataset.classes[labels]))

   
```

```python
with open("output.txt", 'w') as f:
  for line in output.items():
        f.write(str(line))
        f.write('\n')
```
